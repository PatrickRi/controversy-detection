```python
import networkx as nx
import matplotlib.pyplot as plt

```

```python
g1 = nx.complete_graph(range(10))
g2 = nx.complete_graph(range(10, 20))
#fig, ax = plt.subplots(1, 1, figsize=(50, 50), dpi=150)
nx.draw_networkx(g1, with_labels=True)
plt.show()
nx.draw_networkx(g2, with_labels=True)
plt.show()
```

```python
g_big = nx.compose(g1, g2)
g_big.add_edge(1, 10)
nx.draw_networkx(g_big, with_labels=True)
plt.show()

```
