```python
import sys,os,os.path
os.environ['METIS_DLL']='C:\\Users\\Patrick\\Documents\\TU\\2019S\\Thesis_Code\\metis-5.1.0\\build\\windows\\libmetis\\Release\\metis.dll'
import networkx as nx
import metis
import matplotlib.pyplot as plt
```

```python
G = metis.example_networkx()
(edgecuts, parts) = metis.part_graph(G, 2)
colors = ['red','blue','green']
color_map = []
for i, p in enumerate(parts):
    color_map.append(colors[p])
```

```python
nx.draw(G, node_color=color_map, with_labels=True)
plt.show()
```

```python
G = nx.read_gml("karate.gml", label='id')
(edgecuts, parts) = metis.part_graph(G, 2)
colors = ['red','blue','green']
color_map = []
left = []
right = []
for i, p in enumerate(parts):
    if p == 0:
        left.append(list(G.nodes)[i])
    else: 
        right.append(list(G.nodes)[i])
    color_map.append(colors[p])
```

```python
nx.draw(G, node_color=color_map, with_labels=True)
plt.show()
```

```python
fleft = open('karate_left.txt', 'w+')
for i in left:
    fleft.write(str(i)+'\n')
fleft.close()

fright = open('karate_right.txt', 'w+')
for i in right:
    fright.write(str(i)+'\n')
fright.close()
```
