```python
import sys,os,os.path
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
```

```python
# f3 = open("force_directed/" + file2 + "_positions.txt")
f3 = open("code/force_directed/polblogs_positions.txt")
G = nx.read_gml('polblogs.gml', label='id')
# f3 = open("LinLogLayout/" + file2 + "_positions.txt")
lines3 = f3.readlines()

dict_positions = {}
for i in range(len(lines3)):
    line1 = lines3[i].strip()
    line1_split = line1.split("\t")
    node = int(line1_split[0])
    [x, y] = [float(line1_split[1].split(",")[0]), float(line1_split[1].split(",")[1])]
    dict_positions[node] = [x, y]
fig, ax = plt.subplots(1, 1, figsize=(50, 50), dpi=150)
nx.draw_networkx(G, pos=dict_positions, node_size=5, with_labels=True, ax=ax)
plt.show()
```
